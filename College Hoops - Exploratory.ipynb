```python
##Packages
from sportsreference.ncaab.schedule import Schedule
from sportsreference.ncaab.teams import Teams
import pandas
import numpy
pandas.set_option('display.max_columns', 500)
pandas.set_option('display.max_rows', 500)
```

```python
##Gets list of all team names
team_list = []
teams = Teams()

for team in teams:
    abb = team.dataframe
    abb = abb['abbreviation'].tolist()
    abb = team_list.extend(abb)
team_list
```

```python
##VERY EXPENSIVE - roughly one minute per team
##Loops through the list of all teams and brings back box score details for all games
master = pandas.DataFrame()

for team in team_list:
    results = Schedule(team,year='2020').dataframe_extended
    master = pandas.concat([master,results],axis=0,sort=True)
master.to_csv('ncaab_box_scores.csv')
```
